```python
from IPython.display import Markdown
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

train = pd.read_csv('data/train_data.csv')
test = pd.read_csv('data/test_data.csv')

# Pairgrid for train data for climaticZone
hue = 'climaticZone'
vars_per_line = 5
all_vars = list(train.columns.symmetric_difference([hue]))

for var in all_vars:
    rest_vars = list(all_vars)
    rest_vars.remove(var)
    display(Markdown(f"## {var}"))
    while rest_vars:
        line_vars = rest_vars[:vars_per_line]
        del rest_vars[:vars_per_line]
        line_var_names = ", ".join(line_vars)
        display(Markdown(f"### {var} vs {line_var_names}"))
        sns.pairplot(train, x_vars=line_vars, y_vars=[var], hue=hue, palette='bright', )
        plt.show()
        plt.close()
```
